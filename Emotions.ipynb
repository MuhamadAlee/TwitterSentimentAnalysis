```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize

def DatasetModel(tweet):
    
    phrase=[]
    text_tokens = word_tokenize(tweet)
    tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]
    for item in tokens_without_sw:
        phrase.append(lemmatizer.lemmatize(porter.stem(item)))

    
    
    
    angerCounter=0
    anticipationCounter=0
    disgustCounter=0
    fearCounter=0
    joyCounter=0
    negativeCounter=0
    positiveCounter=0
    sadnessCounter=0
    surpriseCounter=0
    trustCounter=0
    arr=[]
    for i in range(11):
        if i==0: 
            arr=anger
        elif i==1:
            arr=anticipation
        elif i==2:
            
            arr=disgust
        elif i==3:
            
            arr=fear
        elif i==4:
            
            arr=joy
        elif i==5:
            
            arr=negative
        elif i==6:
            
            arr=positive
        elif i==7:
            
            arr=sadness
        elif i==8:
            
            arr=surprise
        elif i==9:
            
            arr=trust
     
        for item in phrase:
            for wd in arr:
                if item==wd:
                    if i==0:
                        angerCounter+=1
                    elif i==1:
                        anticipationCounter+=1
                    elif i==2:
                        disgustCounter+=1
                    elif i==3:
                        fearCounter+=1
                    elif i==4:
                        joyCounter+=1
                    elif i==5:
                        negativeCounter+=1
                    elif i==6:
                        positiveCounter+=1
                    elif i==7:
                        sadnessCounter+=1
                    elif i==8:
                        surpriseCounter+=1
                    elif i==9:
                        trustCounter+=1
                    





    highest=max(angerCounter,anticipationCounter,disgustCounter,fearCounter,joyCounter,negativeCounter,positiveCounter,sadnessCounter,surpriseCounter,trustCounter)

    if highest==angerCounter and highest!=0:
        return "Anger"
    elif highest==anticipationCounter and highest!=0:
        return "Anticipation"
    elif highest==disgustCounter and highest!=0:
        return "Disgust"
    elif highest==fearCounter and highest!=0:
        return "Fear"
    elif highest==joyCounter and highest!=0:
        return "Joy"
    elif highest==negativeCounter and highest!=0:
        return "Neagtive"
    elif highest==positiveCounter and highest!=0:
        return "Positive"
    elif highest==sadnessCounter and highest!=0:
        return "Sadness"
    elif highest==surpriseCounter and highest!=0:
        return "Surprise"
    elif highest==trustCounter and highest!=0:
        return "Trust"
    else:
        return "Neutral"


```

```python
datafr['Emotion'] = datafr['Tweets'].apply(DatasetModel)
```
